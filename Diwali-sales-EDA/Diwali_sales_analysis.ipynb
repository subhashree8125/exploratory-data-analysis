4.which product category has more profit margin?

````python


# Group by Product Category and calculate average Amount
grouped = df.groupby('Product_Category')['Amount'].mean()
print(grouped)
# Find the top category
top = grouped.idxmax()
print(f"\nâ Category with highest average amount: {top}")

# Plot the average amount by category
plt.figure(figsize=(10,6))
grouped.sort_values().plot(kind='bar', color='skyblue')
plt.title('Average Amount by Product Category')
plt.xlabel('Average Amount')
plt.ylabel('Product Category')

# Highlight the top category

plt.legend()
plt.tight_layout()
plt.show()
````

5.which product category has more revenue generated?

````python

# Group by Product Category and sum the Amount (Revenue)
revenue_by_category = df.groupby('Product_Category')['Amount'].sum()

# Print all revenues
print("\nRevenue generated by each Product Category:\n")
print(revenue_by_category)

# Find and print the top category
top_category = revenue_by_category.idxmax()
top_revenue = revenue_by_category.max()
print(f"\nâ Category with the highest revenue: {top_category} ({top_revenue:.2f})")

# Plotting
plt.figure(figsize=(10, 6))
revenue_by_category.sort_values().plot(kind='barh', color='lightgreen')
plt.title('Revenue by Product Category')
plt.xlabel('Total Revenue')
plt.ylabel('Product Category')

# Highlight top category with a line
plt.legend()
plt.tight_layout()
plt.show()
````

6.Which state has more orders?

````python

# Count number of orders per state (assuming each row is one order)
orders_by_state = df['State'].value_counts().sort_values(ascending=True)

# Print the sorted order counts
print("\nNumber of Orders by State (Highest to Lowest):\n")
print(orders_by_state)

# Plotting
plt.figure(figsize=(10, 6))
orders_by_state.plot(kind='bar', color='lightcoral')
plt.title('Number of Orders by State')
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

7.Which marital status has given more sales?

````python


# Group by Marital_Status and sum the Amount
sales_by_marital_status = df.groupby('Marital_Status')['Amount'].sum()

# Print the result
print("\nTotal Sales by Marital Status:\n")
print(sales_by_marital_status)

# Plotting
plt.figure(figsize=(6, 4))
sales_by_marital_status.plot(kind='pie', color=["lightgreen","lightcoral"])
plt.title('Total Sales by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

8.Maximum Sales is happening from which state?






````python

# Group by State and sum the Amount
sales_by_state = df.groupby('State')['Amount'].sum()

# Print all states and their total sales
print("\nTotal Sales by State:\n")
print(sales_by_state.sort_values(ascending=False))

# Find the state with the highest sales
top_state = sales_by_state.idxmax()
top_sales = sales_by_state.max()
print(f"\nâ State with the maximum sales: {top_state} ({top_sales:.2f})")

plt.figure(figsize=(10, 6))
sales_by_state.sort_values().plot(kind='bar', color='mediumorchid')
plt.title('Total Sales by State')
plt.xlabel('Total Sales Amount')
plt.ylabel('State')
plt.tight_layout()
plt.show()

````

````python


# Create a pivot table of counts (Age Group vs Product Category)
heatmap_data = df.pivot_table(index='Age Group', columns='Product_Category', values='Product_ID', aggfunc='count', fill_value=0)
print(heatmap_data)
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlGnBu')

# Customizing the heatmap
plt.title('Product Categories Bought by Age Group (Heatmap)')
plt.xlabel('Product Category')
plt.ylabel('Age Group')
plt.tight_layout()
plt.show()
````
